```python
import pandas as pd
from pathlib import Path

DATA_RAW = Path("../../data/raw")
pd.set_option("display.max_columns", 50)
pd.set_option("display.width", 200)
```

```python
path = DATA_RAW / "statcan" / "statcan_retail_trade_food.csv"
print(f"File size: {path.stat().st_size / 1e6:.1f} MB")
df = pd.read_csv(path)
print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
df.head()
```

```python
path = DATA_RAW / "statcan" / "statcan_food_cpi.csv"
print(f"File size: {path.stat().st_size / 1e6:.1f} MB")
df = pd.read_csv(path)
print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
df.head()
```

```python
path = DATA_RAW / "openfoodfacts" / "en.openfoodfacts.org.products.csv.gz"
print(f"File size: {path.stat().st_size / 1e6:.1f} MB")

# Read only first 1000 rows to inspect structure — full file is huge
df = pd.read_csv(
    path,
    sep="\t",
    compression="gzip",
    nrows=1000,
    low_memory=False,
    on_bad_lines="skip",
)
print(f"Sample shape: {df.shape}")
print(f"Total columns: {len(df.columns)}")
print(f"First 30 columns: {df.columns.tolist()[:30]}")
df.head(3)
```

```python
path = DATA_RAW / "dunnhumby" / "transaction_data.csv"
print(f"File size: {path.stat().st_size / 1e6:.1f} MB")
df = pd.read_csv(path, nrows=10000)
print(f"Sample shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
df.head()
```

```python
path = DATA_RAW / "dunnhumby" / "product.csv"
print(f"File size: {path.stat().st_size / 1e6:.1f} MB")
df = pd.read_csv(path)
print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
print(f"\nDepartments: {df['DEPARTMENT'].unique()}")
df.head()
```
